# EDA: NASA C-MAPSS Turbofan Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# FD001 subset
data = pd.read_csv('../data/raw/train_FD001.txt', sep='\s+', header=None)
cols = ['engine_no', 'cycle'] + [f'op_setting_{i}' for i in range(1,4)] + [f'sensor_{i}' for i in range(1,22)]
data.columns = cols
data.head()
```

## Plot RUL (manually computed)

```python
max_cycles = data.groupby('engine_no')['cycle'].transform('max')
data['RUL'] = max_cycles - data['cycle']
# Show a few engines
data[data['engine_no'] < 3].plot(x='cycle', y='RUL', title='RUL Degrading for Engines 1-2')
plt.show()
```

## Sensor Trends

```python
# Plot sensor 2 for engine #1
e1 = data[data['engine_no'] == 1]
plt.plot(e1['cycle'], e1['sensor_2'])
plt.title('Sensor 2 over time (engine 1)')
plt.xlabel('cycle')
plt.ylabel('sensor_2 reading')
plt.show()
```
